# EXCEPTION HANDLING

#### Need For Exception Handling


We Need Exception Handling When:-
- The Program terminates whenever an error occurs
- Sudden Termination can corrupt the program
- Exception may cause data loss

#### What is Exception Handling?

- An Exception is a runtime error that terminates the execution of a program
- Exception Handling generates a message in correspondance to the occurence of an error and takes the right step to handle it

#### Types Of Errors

In General there are three types of Errors:-
- Compile Time Errors
- Logical Errors
- Runtime Errors

#### How Exception Handling Works?

- You use a try block, in which you write the code that needs to be run.
- You put an except keyword where you throw an error message
- If there is no exception you use else block
- Finally Block which will always be executed if there is exception or no exception

##### Steps
- You start with a try block 
- If exception occurs you use except block and then the finally block.
- But if no exception occurs you use else block and then the finally block


##### Points To Remember

> You can write try block without any except blocks

> You can write several except blocks for a single try block

> You can write multiple except blocks to handle multiple exceptions

> You can write exception as an object

> You can write multiple exception within tuple

#### Divide By Zero Error

```python
a = 10
b = 0
c = a/b
print("a/b = %d" %c)
```

###### Solution 

```python
try:
    a = int(input("Enter a:"))
    b = int(input("Enter b:"))
    c = a/b
except:
    print("Can't Divide By Zero")
```

###### Alternative

```python
a = 70
b = 0
try:
    print(a/b)
    print("New Line")
except ZeroDivisionError:
    print("Do not divide number by zero")
print("Completed")
```

###### Except Block doesnt get executed as there was no exception in the below code

```python
a = 70
b = 9
try:
    print(a/b)
    print("New Line")
except ZeroDivisionError:
    print("Do not divide number by zero")
print("Completed")
```

###### In this alternative we print an exception object

```python
try:
    a = int(input("Enter a:"))
    b = int(input("Enter b:"))
    c = a/b
    print("a/b = %d"%c)
except Exception:
    print("Can't Divide By Zero")
    print(Exception)
else:
    print("Completed")
```
